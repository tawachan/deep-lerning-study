```python
import pandas as pd
```

```python
df = pd.read_csv("./data/all_currencies.csv", index_col=0)
df = df.fillna(0)
df.head()
```

```python
df.dtypes
```

```python
# df = df.drop("Market Cap", axis=1)
```

```python
dates = pd.to_datetime(df["Date"], format="%Y-%m-%d")
df["Year"] = dates.dt.year
df["Month"] = dates.dt.month
df["Day"] = dates.dt.month
df["DayOfWeek"] = dates.dt.dayofweek
```

```python
df.to_csv("./data/preprocessed_data.csv")
```

```python
btc = df[df["Symbol"] == "BTC"]
btc.to_csv("./data/preprocessed_data_btc.csv")

btc_next = btc.shift(-1)[:-1]
btc_cur = btc[:-1]

btc_diff = btc_cur.copy()[["Symbol", "Date", "Volume", "Year", "Month", "Day", "DayOfWeek"]]
btc_diff["OpenDiff"] = btc_cur["Open"] - btc_cur["Open"]
btc_diff["HighDiff"] = btc_next["High"] - btc_cur["High"]
btc_diff["LowDiff"] = btc_next["Low"] - btc_cur["Low"]
btc_diff["CloseDiff"] = btc_next["Close"] - btc_cur["Close"]
btc_diff["VolumeDiff"] = btc_next["Volume"] - btc_cur["Volume"]
btc_diff.to_csv("./data/btc_diff.csv")
btc_diff
```
